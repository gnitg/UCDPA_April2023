# Visualisation

Note: This will use datasets from DataCamp (Summer Olympics 2016 and Austin Weather)

## Importing Modules

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Scatter Plot

```python
summer2016 = pd.read_csv('data/summer2016.csv')
summer2016.head()
```

### Pandas

```python
summer2016.plot.scatter('Height','Weight', s=2)
```

### Matplotlib

```python
plt.scatter(summer2016['Height'],summer2016['Weight'], s=2)
plt.show()
```

### Seaborn

```python
sns.scatterplot(x='Height',y='Weight', data=summer2016, hue='Medal', size='Age')
```

```python
markers = {"M": "s", "F": "X"}
hue = {"M": "blue", "F": "red"}
sns.scatterplot(x='Height',y='Weight', data=summer2016, 
                style='Sex', hue='Sex', palette=hue, size='Age', markers=markers)
```

## Line Plot

```python
austin = pd.read_csv('data/austin_weather.csv')
austin.head()
```

### Pandas

```python
austin.plot.line(y='MLY-CLDD-BASE45')
```

### Matplotlib

```python
plt.subplots(figsize=(10,5))

plt.plot(austin['DATE'],austin['MLY-CLDD-BASE45'])
plt.plot(austin['DATE'],austin['MLY-CLDD-BASE50'])

plt.xlabel('Date (months)')
plt.ylabel('Amount of Rain (mm)')
plt.title('Rainfall across Year')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Pandas

If we try to plot the data as bars, it will be a mess, as every point is getting its own bar. We will need to aggregate it first by grouping it.

```python
tips.plot.bar(x='smoker', y='tip')
```

```python
tips.groupby('smoker').mean()['total_bill'].plot.bar()
```

```python
tips_grp = tips.groupby('smoker').mean()['total_bill']
```

```python
tips_grp
```

```python
plt.bar(tips_grp.index, tips_grp.values)
```

Fortunately, Seaborn groups and aggregates the data automatically.

```python
sns.barplot(data=tips, y='smoker', x='total_bill', hue='sex')
```

## Count Plot

```python
titanic = sns.load_dataset('titanic')
```

```python
values = titanic['sex'].value_counts().values
sex = titanic['sex'].value_counts().index
```

```python
plt.bar(sex, values)
```

```python
sns.countplot(data=titanic, x='sex', hue='class')
```
